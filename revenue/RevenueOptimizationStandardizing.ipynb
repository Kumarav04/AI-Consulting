```python
# Script to standardize the CSVs to the acceptable format
# The acceptable column names are: Date, Product_ID, Product_name, Units_sold, Units_price, Revenue, Profit(optional)
import pandas as pd
```

```python
input_csv = input(str('Enter the CSV: '))
```

```python
df = pd.read_csv(input_csv)
df = df.dropna()
df
```

```python
rename_mapping = {
    'Order ID' : 'Order_ID',
    'Customer ID' : 'Order_ID',
    'Date': 'Order_Date',
    'Order Date' : 'Order_Date',
    'Product' : 'Product_Name',
    'Category' : 'Product_Name',
    'Item Type' : 'Product_Name',
    'Product Category' : 'Product_Name',
    'Quantity Ordered' : 'Units_Sold',
    'Units Sold' : 'Units_Sold',
    'Qty' : 'Units_Sold',
    'Quantity' : 'Units_Sold',
    'Price Each' : 'Unit_Price',
    'Unit Price' : 'Unit_Price',
    'Price per Unit' : 'Unit_Price',          
}
```

```python
df = df.rename(columns=rename_mapping)
df
```

```python
order = ['Order_ID', 'Order_Date', 'Product_Name', 'Units_Sold', 'Unit_Price']
```

```python
df = df[order]
```

```python
df
```

```python
df['Units_Sold'] = pd.to_numeric(df['Units_Sold'], errors='coerce')
df['Unit_Price'] = pd.to_numeric(df['Unit_Price'], errors='coerce')
```

```python
df['Revenue'] = df['Units_Sold']*df['Unit_Price']
```

```python
df
```

```python
standardized_df = df.to_csv('standardized'+input_csv)
```
